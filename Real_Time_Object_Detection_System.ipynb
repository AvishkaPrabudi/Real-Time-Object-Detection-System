```python
from ultralytics import YOLO

# Load pretrained model
model = YOLO("yolov8n.pt")

# Train
model.train(
    data="coco128.yaml",
    epochs=10,     # demo ekata 10 enough
    imgsz=640,
    batch=16,
    name="custom_yolo_model"
)
```

```python
metrics = model.val()
print(metrics)
```

```python
results = model.predict(
    source="https://ultralytics.com/images/bus.jpg",
    save=True,
    conf=0.25
)
```
